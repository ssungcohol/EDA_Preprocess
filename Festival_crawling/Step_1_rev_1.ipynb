```python
from selenium import webdriver
from selenium.webdriver.common.by import By
# from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
import csv
```

```python
url = 'https://www.visitjeju.net/kr/festival/list?menuId=DOM_000001718007000000&cate1cd=cate0000001360#p1&month='
driver = webdriver.Chrome()
driver.get(url)

# csv 파일 생성
with open(r'C:\Users\sjhj8\Desktop\Golden_Planet\Festival\Crawling\step_1.csv', 'w', encoding = 'utf-8-sig', newline = '') as f:
    csvWriter = csv.writer(f)
    csvWriter.writerow(['축제명', '기간', '월', '위치', '세부지역'])

    page = 12

    time.sleep(1)

    # 페이지
    for i in range(0, page + 1):

        if i == 0 or i == 10 or i == 11 or i == 12:
            m = i
        else:
            m = '0' + str(i)
        
        driver.get(f'https://www.visitjeju.net/kr/festival/list?menuId=DOM_000001718007000000&cate1cd=cate0000001360#p1&month={m}')
        print(f'Page {i} loaded')

        time.sleep(1)
        
        event_lists = driver.find_elements(By.CSS_SELECTOR, '.event_list.clear')
        print('event_list')
        
        for event_list in event_lists:
            print('Event list crawling start')
    
            lis = event_list.find_elements(By.CSS_SELECTOR, 'li')
    
            for li in lis:
                try:
                    title = li.find_element(By.CLASS_NAME, 'item_tit').text
                    period = li.find_element(By.CLASS_NAME, 'item_period').text
                    month = i
                    location = li.find_element(By.CLASS_NAME, 'item_location').text.split(">")[0]
                    detail_location = li.find_element(By.CLASS_NAME, 'item_location').text.split(">")[1]
                
                except Exception:
                    print('--------exception--------')
                    continue
    
                print(title, month, detail_location)
                csvWriter.writerow([title, period, month, location, detail_location])

driver.quit()
# f.close()
```

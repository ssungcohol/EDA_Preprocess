```python
# 빈 데이터프레임 준비
# raw = pd.DataFrame()

data_frames = []

# Merge
for file in os.listdir(r'C:\jupyter_projects\ITS\old_data_2'):
    fpath = os.path.join(file_path, file)
    print(fpath)
    try:
        temp = pd.read_csv(fpath)
        data_frames.append(temp)
    except FileNotFoundError:
        print(f"File not found : {fpath}")
    except pd.errors.EmptyDataError:
        print(f"Empty data file : {fpath}")
    except Exception as e:
        print(f"An Error occurred : {e}")

# 데이터 프레임 리스트를 하나로 병합
if data_frames :
    raw = pd.concat(data_frames, ignore_index = True)
    print("Merge completed")
else:
    raw = pd.DataFrame()
    print("No file to merge")

raw.to_csv(r'C:\jupyter_projects\ITS\old_data_merge_2\old_data_merge.csv', index=False, encoding = 'utf-8-sig')

print(raw)

raw.info
```

```python
import pandas as pd
import os
```

```python
df = pd.read_csv(r'C:\jupyter_projects\ITS\old_data_merge_2\old_data_merge.csv', encoding = 'utf-8-sig')
df.head()
```

```python
df_season = df.copy()
df_season.head()
```

### '월' 데이터 추출

```python
df_season['월'] = df_season['돌발일시'].str.slice(5, 7).astype(int)
```

```python
df_season.head()
```

### '계절' 데이터 추출

```python
df_season['계절'] = (df_season['월'] // 3) % 4

season_mapping = {0 : '겨울', 1 : '봄', 2 : '여름', 3 : '가을'}

df_season['계절'] = df_season['계절'].map(season_mapping)
```

```python
df_season.head()
```

### 돌발 상황 발생 '시간대' 데이터 추출

```python
df_season['상황 발생 시간대'] = df_season['돌발시작일시'].str.slice(11, 13)
df_season['상황 발생 시간대'] = df_season['상황 발생 시간대'].str.replace(':', '')
df_season['상황 발생 시간대'] = df_season['상황 발생 시간대'].astype(int)
df_season['상황 발생 시간대'] = (df_season['상황 발생 시간대'] // 6) % 4

time_zone_mapping = {0 : '새벽', 1 : '오전', 2 : '오후', 3 : '저녁'}

df_season['상황 발생 시간대'] = df_season['상황 발생 시간대'].map(time_zone_mapping)

df_season.head()
```

```python
df_season['돌발구분'] = df_season['돌발구분'].astype('int')
df_season.info()
```

### 도로명 추출

```python
def extract_road_name(text):
    road_name = text.split()
    if '>' in road_name[0]:
        return road_name[1]
    return road_name[0]

def road_name(df):
    df_season['돌발내용'] = df_season['돌발내용'].astype(str)
    df_season['도로명'] = df_season['돌발내용'].apply(extract_road_name)
    return df_season
```

```python
df_season = road_name(df_season)

df_season.head()
```

### 제주시 / 서귀포시 구분

```python
import chardet

# 파일 인코딩 확인
def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        raw_data = f.read()
    result = chardet.detect(raw_data)
    return result['encoding']
   

folder_path = r'C:\jupyter_projects\ITS\jeju_road_name'

for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    print(f"Processing file : {file_path}")

     # 파일 인코딩 형식 확인
    encoding = detect_encoding(file_path)

    print(encoding)
```

```python
df_jeju = pd.read_csv(r'C:\jupyter_projects\ITS\jeju_road_name\jeju_2.csv', encoding = encoding)
df_seogwipo = pd.read_csv(r'C:\jupyter_projects\ITS\jeju_road_name\seogwipo_2.csv', encoding = encoding)

# road_jeju = df_jeju['도로명주소'].str.split()
# road_seogwipo = df_seogwipo['도로명주소'].str.split()

road_jeju = df_jeju['도로명']
road_seogwipo = df_seogwipo['도로명']
```

```python
road_name_jeju = []
road_name_seogwipo = []

for road_index in range(len(road_jeju)):
    if "로" in road_jeju[road_index][1] or "길" in road_jeju[road_index][1]:
        road_name_jeju.append(road_jeju[road_index][1])
    else:
        road_name_jeju.append(road_jeju[road_index][2])

for road_index in range(len(road_seogwipo)):
    if "로" in road_seogwipo[road_index][1] or "길" in road_seogwipo[road_index][1]:
        road_name_seogwipo.append(road_seogwipo[road_index][1])
    else:
        road_name_seogwipo.append(road_seogwipo[road_index][2])

road_name_jeju = set(road_name_jeju)

road_name_seogwipo = set(road_name_seogwipo)
```

```python
jeju_2 = df_jeju['도로명']
jeju_2 = jeju_2.values
jeju_2 = jeju_2.tolist()

seogwipo_2 = df_seogwipo['도로명']
seogwipo_2 = seogwipo_2.values
seogwipo_2 = seogwipo_2.tolist()
```

```python
# 사용 X
def seperate_jeju(df_season, road_name_jeju, road_name_seogwipo):

    df['지역구분'] = None

    for idx, row in df_season.iterrows():
        if row['도로명'] in road_name_jeju:
            df_season.at[idx, '지역구분'] = '제주시'
        elif row['도로명'] in road_name_seogwipo:
            df_season.at[idx, '지역구분'] = '서귀포시'
        else:
            df_season.at[idx, '지역구분'] = row['도로명']

    return df_season


def seperate_jeju_2(df_season, jeju_2, seogwipo_2):

    df['지역구분'] = None

    for idx, row in df_season.iterrows():
        if row['도로명'] in jeju_2:
            df_season.at[idx, '지역구분'] = '제주시'
        elif row['도로명'] in seogwipo_2:
            df_season.at[idx, '지역구분'] = '서귀포시'
        else:
            df_season.at[idx, '지역구분'] = row['도로명']

    return df_season
```

```python
# '월', '계절', '도로명', '지역구분' column까지 파일 저장

df_season = seperate_jeju_2(df_season, jeju_2, seogwipo_2)

df_season

save_folder_path = r'C:\jupyter_projects\ITS\old_data_merge_2'
save_path = os.path.join(save_folder_path, filename)


df_season.to_csv('C:\\jupyter_projects\\ITS\\old_data_merge_2\\old_data_merge_fin.csv', index = False, header = True, encoding = encoding)
```

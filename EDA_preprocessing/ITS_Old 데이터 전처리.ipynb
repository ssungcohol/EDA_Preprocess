```python
# 파일 다운로드 함수
def download(url, file_name):
    with open(file_name, "wb") as file:   # open in binary mode
        response = get(url)               # get request
        file.write(response.content)
```

```python
from datetime import date, timedelta

# 시작 날짜와 끝 날짜 정의
start_date = date(2021, 1, 1)
end_date = date(2022, 11, 21)

# 날짜들을 담을 리스트 초기화
date_list = []

# 시작 날짜부터 끝 날짜까지 반복하며 리스트에 추가
current_date = start_date
while current_date <= end_date:
    # 날짜를 'YYYYMMDD' 포맷의 문자열로 변환
    date_str = current_date.strftime('%Y%m%d')
    # '.zip'를 붙여 리스트에 추가
    date_list.append(date_str + '.zip')
    # 다음 날짜로 이동
    current_date += timedelta(days=1)

# 결과 출력 (테스트 출력)
print(date_list)
```

```python
for i in date_list:
    new_url = url + i[:4] + '/' + i
    download(new_url, i)
```

### 파일 압축 해제

```python
import zipfile

def extract_files_in_directory(directory):
    # 디렉토리 내의 모든 파일을 선택
    for entry in os.listdir(directory):
        entry_path = os.path.join(directory, entry)
        
        # 파일인지 확인
        if os.path.isfile(entry_path):
            # .zip 파일 압축 해제
            if entry_path.endswith('.zip'):
                try:
                    with zipfile.ZipFile(entry_path, 'r') as zip_ref:
                        extract_path = os.path.splitext(entry_path)[0]
                        os.makedirs(extract_path, exist_ok=True)
                        zip_ref.extractall(extract_path)
                        print(f'Extracted: {entry_path} -> {extract_path}')
                except :
                    print(f'Unvailable: {entry_path}')
                    pass
```

### 제주 위·경도 필터링

```python
import csv
import chardet
```

```python
# 파일 인코딩 확인
def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        raw_data = f.read()
    result = chardet.detect(raw_data)
    return result['encoding']


# 위·경도 설정
def filter_location(df):
    # 위도
    latitude = (df.iloc[:, -1] >= 33.1127000) & (df.iloc[:, -1] <= 33.6000000)
    # 경도
    longitude = (df.iloc[:, -2] >= 126.0843000) & (df.iloc[:, -2] <= 126.9724370)

    return df[latitude & longitude]
```

```python
# 폴더 경로 설정 (원하는 폴더 경로 입력)
folder_path = r'C:\jupyter_projects\ITS\decompact'

# 변경하고자 하는 인코딩 Typer 입력
encoding_type = 'utf-8-sig'

# 저장 폴더 경로 설정
save_folder_path = r'C:\jupyter_projects\ITS\filtering_data_2'

# 폴더 내의 모든 파일에 대해 반복
for filename in os.listdir(folder_path):

    # csv 파일 필터링
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)

        # 파일 별 인코딩 형식이 다를 수 있음으로, 인코딩 형식 확인
        encoding = detect_encoding(file_path)
        print(f"Detect encoding type : {encoding}")

        try:
            df = pd.read_csv(file_path, header = None, encoding = encoding, on_bad_lines = 'skip')
        except pd.errors.ParserError as e:
            print(f"Error parsing {filename} : {e}")
            continue

        filtered_df = filter_location(df)

        save_path = os.path.join(save_folder_path, filename)

        filtered_df.to_csv(save_path, index = False, header = False, encoding = encoding_type)

        print(f'{filename} is filtering complete.')

print('')
print('=' * 40)
print(f'All file filtering clear')
print(f'All file encoding type : {encoding_type}')
```

### (상황처리)소요시간 & 차로 데이터 추출

```python
# (상황처리) 소요시간
def lead_time(df):
    # 날짜 형식 지정
    df['돌발시작일시'] = pd.to_datetime(df['돌발시작일시'])
    df['돌발종료일시'] = pd.to_datetime(df['돌발종료일시'])
    
    # 소요시간 column 생성, date 입력 및 column 위치 지정
    df.insert(4, '소요시간', df['돌발종료일시'] - df['돌발시작일시'])

    return df

# 차로 Data 추출
def extract_lane(text):
    road_index = text.rfind('차로')
    if road_index is None:
        return '0'
    return text[max(road_index - 2, 0) : road_index].lstrip()

# 차로 column 추가
def road_index(df):
    df['돌발내용'] = df['돌발내용'].astype(str)
    df['차로'] = df['돌발내용'].apply(extract_lane)
    return df
```

```python
# 폴더 경로 설정 (원하는 폴더 경로 입력)
folder_path = r'C:\jupyter_projects\ITS\filtering_data_2'

# 저장 폴더 경로 설정
save_folder_path = r'C:\jupyter_projects\ITS\old_data_2'

# header 없을 시 생성하고자 하는 header 명 입력
column_headers = ['돌발일시', '??코드', '돌발시작일시', '돌발종료일시', '돌발구분', '돌발상세구분', '돌발내용', '위도', '경도']

# 원하는 인코딩 Type 설정
encoding_type = 'utf-8-sig'

# 폴더 내의 모든 파일에 대해 반복
for filename in os.listdir(folder_path):

    # csv 파일 filtering
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        print(f"Processing file : {file_path}")

    # 파일 인코딩 형식 확인
    encoding = detect_encoding(file_path)

    try:
        df = pd.read_csv(file_path, header=None, encoding = encoding, on_bad_lines = 'skip')
    except pd.errors.EmptyDataError:
        print(f"EmptyDataError : No columns to parse from file {filename}")
        continue
    except pd.errors.ParserError as e:
        print(f"Error parsing {filename} : {e}")
        continue

    # 파일이 비어있는 경우 (No data) 스킵
    if df.empty:
        print(f"Warning : {filename} is empty.")
        continue

    # header 추가
    if len(df.columns) == len(column_headers):
        df.columns = column_headers
    else:
        print(f"Warning : Number of columns in file {filename} does not match the number of columns you wnat to enter.")
        continue

    df = lead_time(df)

    df = road_index(df)

    df = df.reindex(['돌발일시', '??코드', '돌발시작일시', '돌발종료일시', '소요시간', '돌발구분', '돌발상세구분', '돌발내용', '차로', '위도', '경도'], axis = 1)

    save_path = os.path.join(save_folder_path, filename)

    df.to_csv(save_path, index = False, header = True, encoding = encoding_type)

    print(f"{filename} lead_time, road_index extract complete.")

print("All file extracting complete.")
```
